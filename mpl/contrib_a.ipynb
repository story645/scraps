```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cdf  = pd.read_csv("contributors.csv")
```

```python
cdf.info()
```

```python
cdf['contributions'].describe()
```

```python
# distribution of contributor commit counts 
%matplotlib inline
import seaborn as sns
_ = sns.distplot(cdf['contributions'])
```

```python
bins = [0, 1, 10, 20, 30, 50, 60, 70, 80, 90, 100, 5000]
cdf['contributions'].value_counts(bins=bins, sort=False)
fig, ax = plt.subplots()
_ = cdf['contributions'].value_counts(bins=bins, sort=False).plot.bar(ax=ax)
_ = ax.set_ylabel("# of contributors")
_ = ax.set_xlabel("# of contributions")
```

```python
cone = cdf[cdf['contributions']>20][['login', 'contributions']]
```

```python
fig, ax = plt.subplots(figsize=(15,5))
_ = ax.set_title("Contributors with over 50 commits")
_ = ax.set_yscale('log')
_ = cone[['contributions']].plot.bar(ax=ax)
_= ax.set_xticklabels(cone['login'], rotation=90)
```
